# Aggregation Simulation

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random
```

Plots

```python
s = np.random.zipf(1.2, 1000)
def visualize_trace(trace, hits=[]):
    fig = plt.figure(figsize=(20, 4))
    # TODO: it would be nice to size the markers so that they don't overlap.
    hits = hits or [False for _ in range(len(s))]
    hits = ["green" if x else "red" for x in hits]
    plt.scatter(range(len(s)), s, marker='s', color=hits, s=4)
    plt.axis(xmin=0, xmax=len(s), ymin=-50, ymax=500)
    plt.show()
visualize_trace(s)
```

```python
class EvictionPolicy(object):
    def __init__(self):
        pass
    
    def choose_victim(self, key, buf):
        # Can return None to bypass the cache.
        pass
    
    def update_stats(self, key, buf):
        pass
    
class RandomEvictionPolicy(EvictionPolicy):
    def choose_victim(self, key, buf):
        return random.choice(buf.keys())

def simulate_aggregation(stream, buffer_size, policy):
    buf = {}
    savings = 0
    evictions = 0
    transmissions = 0
    hits = []
    for key in stream:
        policy.update_stats(key, buf)
        if key in buf:
            buf[key] += 1
            hits.append(True)
        elif len(buf) < buffer_size:
            buf[key] = 0
            hits.append(False)
        else:
            transmissions += 1
            victim = policy.choose_victim(key, buf)
            if victim:
                evictions += 1
                savings += buf[victim]
                del buf[victim]
                buf[key] = 0
                hits.append(False)
            else:  # bypassed cache
                hits.append(False) 
    transmissions += len(buf.keys())
    savings += sum(buf.values())
    num_hits = len([x for x in hits if x])
    return {'savings': savings, 'transmissions': transmissions, 'evictions': evictions, 'tuples_sent': transmissions, 'hit_trace': hits, 'hits': num_hits}
```

The folowing charts visualize hits/misses against the hashtable.  The y-axis shows keys, while the x-axis shows the progress of time.  "Hits" are shown in green and misses in red.

### Optimal Eviction

Belady's Algorithm.

```python
class OptimalEvictionPolicy(EvictionPolicy):
    def __init__(self, stream):
        self._stream_tail = list(stream)
        
    def update_stats(self, key, buf):
        self._stream_tail = self._stream_tail[1:]
        
    def choose_victim(self, key, buf):
        # Order keys by next use
        def next_use(key):
            if key in self._stream_tail:
                return self._stream_tail.index(key)
            else:
                return float("inf")
        victim = sorted(buf.keys(), key=next_use)[-1]
        return victim
```

```python
results = simulate_aggregation(s, 100, OptimalEvictionPolicy(s))
print "Total tuples sent is %i (%i hits)" % (results['tuples_sent'], results['hits'])
visualize_trace(s, results['hit_trace'])
```

### Randomized Eviction

```python
results = simulate_aggregation(s, 100, RandomEvictionPolicy())
print "Total tuples sent is %i (%i hits)" % (results['tuples_sent'], results['hits'])
visualize_trace(s, results['hit_trace'])
```

### Macroblock Statistics

In this strategy, we partition the keyspace into macroblocks and track hit-rates at a per-macroblock granularity.  This is inspired by a scheme from _Johnson, T.L. et al. 1999. Run-time cache bypassing. IEEE Transactions on Computers. 48, 12 (1999), 1338–1354._

In contrast with that system, our "cache" is fully associative, so we first find whether the poorest-performing macroblock is candidate for eviction.

```python
class MacroblockEvictionPolicy(EvictionPolicy):
    def __init__(self, num_macroblocks, fraction=1):
        self._stats = [0 for _ in range(num_macroblocks)]
        self._fraction = fraction
        
    def update_stats(self, key, buf):
        macroblock = hash(key) % len(self._stats)
        if key in buf:
            self._stats[macroblock] += 1
            
    def choose_victim(self, key, buf):
        key_macroblock = hash(key) % len(self._stats)
        self._stats[key_macroblock] = max(self._stats[key_macroblock] - 1, 0)
        key_hit_rate = self._stats[key_macroblock]
        for victim in buf:
            victim_macroblock = hash(victim) % len(self._stats)
            if self._stats[victim_macroblock] < self._fraction * key_hit_rate:
                self._stats[key_macroblock] = self._stats[victim_macroblock]
                self._stats[victim_macroblock] = 0
                return victim
        return None            
```

```python
results = simulate_aggregation(s, 100, MacroblockEvictionPolicy(20, fraction=0))
print "Total tuples sent is %i (%i hits)" % (results['tuples_sent'], results['hits'])
visualize_trace(s, results['hit_trace'])
```
